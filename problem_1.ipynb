```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Problem 1: Create a sine graph that hasx limits from [-2pi,2pi], and y limits as [-1,1].Also print out X and Y axis in bold, you can print it as two more plots of lines.The colors can be decided by you.

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.sin(x)
plt.plot(x, y)
plt.axhline(0, color='black', linewidth=1.5)
plt.axvline(0, color='black', linewidth=1.5)
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-1, 1)
plt.xlabel('X')
plt.ylabel('sin(x)')
plt.title('Sine')
plt.show()
```

Problem 2: Use 3 markers to understand the range between np.random.rand, np.random.randn and np.random.random as shown in the picture below.

```python
np_neg_ones = np.ones(1000)*-1
np_ones = np.ones(1000)
np_r = np.arange(-500,500,1)
```

```python
plt.style.use('bmh')
rng = np.random.RandomState(0)
markers = ['+', 'x','o']
# for marker in markers:
plt.plot(rng.rand(100), rng.rand(100), markers[0],
          label="marker='{0}_rand'".format(markers[0]))

plt.plot(rng.randn(200), rng.randn(200), markers[1],
          label="marker='{0}_randn'".format(markers[1]))

plt.plot(rng.random(150), rng.random(150), markers[2],
          label="marker='{0}_random'".format(markers[2]))
plt.plot(np_ones, np_r)
plt.plot(np_neg_ones, np_r)
plt.plot(np_r, np_ones)
plt.plot(np_r, np_neg_ones)
plt.legend(numpoints=1)
plt.xlim(-5, 5)
plt.ylim(-5,5);
```

Problem 3. Create a unit circle, given that the points are

```python
center = (0, 0) 
radius = 5
theta = np.linspace(0, 2 * np.pi, 100)
x_points = center[0] + radius * np.cos(theta)
y_points = center[1] + radius * np.sin(theta)
plt.plot(x_points, y_points, color='yellow', linewidth=2)
plt.axis('equal')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Circle Plot')
plt.grid()
plt.show()
```
