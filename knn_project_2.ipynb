```python
X = train[['LotArea', 'BedroomAbvGr', 'FullBath', 'YearBuilt']]
y = train['SalePrice']
```

Model Building: a. Split the dataset into features (X) and the target variable (y). b. Implement a KNN regression model. Start with a reasonable default value for k. c. Use cross-validation (e.g., k-fold cross-validation) to evaluate the model's performance. Choose an appropriate evaluation metric for regression (e.g., mean squared error or mean absolute error).

```python
knn_regressor = KNeighborsRegressor()
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
grid_search = GridSearchCV(knn_regressor, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
```

We built our model with KNN regressor,with amount of parameters to train

Hyperparameter Tuning: a. Implement a grid search or random search to tune the hyperparameters of the KNN regressor. Experiment with different values of k and distance metrics (e.g., Euclidean, Manhattan). b. Determine the best combination of hyperparameters based on the cross-validation results.

```python
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
best_knn_regressor = KNeighborsRegressor(n_neighbors=best_params['n_neighbors'],
                                          weights=best_params['weights'],
                                          p=best_params['p'])
```

```python
best_knn_regressor.fit(X_train, y_train)
y_pred = best_knn_regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```

We have done hyperparameter tuning and found out best parameters from given list

Model Evaluation: a. Train the KNN regressor using the best hyperparameters identified. b. Evaluate the model's performance on a held-out test set using the chosen evaluation metric.
