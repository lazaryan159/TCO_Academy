Write a function that calculates the standard deviation of a pandas dataframe column using only numpy library

```python
def my_std(data):
    mean_values = np.mean(data.values)
    square = (data.values - mean_values) ** 2
    sum_square = np.sum(square)
    variance = sum_square / len(data.values)
    std_dev = np.sqrt(variance)

    return std_dev
```
